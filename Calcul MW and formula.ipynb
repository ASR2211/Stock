```python
!pip install rdkit
'https://www.rdkit.org/docs/GettingStartedInPython.html'
from rdkit import Chem
from rdkit.Chem import rdMolDescriptors, Descriptors

m = Chem.MolFromSmiles('C[C@H](NC(=O)OCc1ccccc1)C(=O)N(C)[C@@H](C)C(=O)N[C@@H](CC(N)=O)C(=O)N[C@@H](CCN(C(=O)CO)[C@@H](c1cc(-c2cc(F)ccc2F)cn1Cc1ccccc1)C(C)(C)C)C(=O)NCCC(=O)N[C@H](CCC(=O)OCc1ccccc1)C(=O)OCc1ccccc1')
display(m)
molecular_formula = rdMolDescriptors.CalcMolFormula(m)
molecular_weight = Descriptors.MolWt(m)
exact_molecular_weight = rdMolDescriptors.CalcExactMolWt(m)
logP = Descriptors.MolLogP(m)

print(f"Molecular Formula: {molecular_formula}")
print(f"Molecular Weight: {molecular_weight}")
print(f"Exact Molecular Weight: {exact_molecular_weight}")
print(f"LogP: {logP}")
```
